```python
import numpy as np
import matplotlib.pyplot as plt
path = "./AFALogo.bmp"
logo = plt.imread(path)
plt.imshow(logo)
plt.gray()
plt.show()
logo.shape
```

Ejercicio 1: cortar las letras dentro del logo

```python
letras = logo[75:160, 20:105] 
  
plt.imshow(letras)
plt.gray()
plt.show()
```

Ejercicio 2: Cortar las estrellas y tranpsonerlas

```python
estrellas = logo[0:40, 20:105]
estrellas_t = estrellas.T
plt.imshow(estrellas_t)
plt.gray()
plt.show()
```

Ejercicio 3: Generar una mascara separando el color de fondo del logotipo

```python
h,w = logo.shape

mask = np.zeros((h,w))
for i in range(h):
    for j in range(w):
        if logo[i][j] != 0:
          mask[i][j] = 1  #lo pongo en blanco

mask
plt.imshow(mask)
plt.gray()
plt.show()
```

Ejercicio 4: Cambiar el color de fondo de blanco a negro

```python
h,w = logo.shape

mask = logo
for i in range(h):
    for j in range(w):
        if logo[i][j] == 1:
          mask[i][j] = 0 


plt.imshow(mask)
plt.gray()
plt.show()
```
